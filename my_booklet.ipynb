# Basic text analysis

```python
# in the following part of basic text analysis we will use 2 texts from nltk books 
from nltk.book import text1, text3
print(f"we will be using \"{text1.name}\" and \"{text3.name}\" for some basic text analysis")
```

```python
# The Text object in itself is not a list nor does it implament itterable 
# but it does have some methods that allow it to behave like a list (indexing and length )
# sometimes called sequence like behaviour 
# It also contains some text analysis methods and some visualization methods 
# The primary data structure for a Text object is a list of strings,
# where each string represents a word or punctuation mark from the text
# sometimes refered to as a "tokized" text (more on toknization later )
from nltk.text import Text

# lexical richness the ratio between the number of unique words 
# to the total number of words in the text
def lexical_richness(text):
    # remmeber Text class implaments indexing and len methods that allow 
    # the method set(text) to work even thou it is NOT an iterable 
    return len(set(text)) / len(text)

print(f"lexical richness of {text1.name} is {lexical_richness(text1)}")
print(f"lexical richness of {text3.name} is {lexical_richness(text3)}")
```

```python
# using some inbuilt text analysis functions of the Text object 
# using the Text.count methoud to assess the importance of a word in the text
def precent_of_text(text,word):
    len_text = len(text)
    given_word_count = text.count(word)
    word_precenteage_of_text = (given_word_count/len_text)*100
    print(f"the word {word} is { word_precenteage_of_text } % of the {text.name} text")

precent_of_text(text1,"God")
precent_of_text(text3,"God")
# so we can conclude tat "God" has a greater importance in 
# the book of genesis then in mobydick 
```

```python
from nltk import FreqDist

# the function FreqDist returns a FreqDist object 
# represesnting the number of occurences of each word in the text
# and providing methods to plot and analyzine this data 
print(type(FreqDist(text1)))
# for example we can use it to plot the cumulative frequence of the 50 most frequent words 
text3FreqDist = FreqDist(text3)
text3FreqDist.plot(50, cumulative=True)
print(f"number of words in the text : {len(text3)}")
```

```python
# we can see that the 50 most common words in the text take 50% of the book 
# we can also use FreqDist to check if there are words in the text that appear only once 
words_that_appear_once = text3FreqDist.hapaxes()
print(f"number of words that appear only once in the text : {len(words_that_appear_once)}")
print(f"example of 10 words that appear only once {words_that_appear_once[:10]}" )
```

```python
# concordance : An alphabetical index of all the words in a text
# or corpus of texts, showing every contextual occurrence of a word. 
# nltk has a built in function for text objects called concordance that 
# prints senteces where a word or a list of words appear  
text3.concordance("god") # all the sentences with the word god 
print("\n\n=====\n\n")
text3.concordance(["god","created"]) # all the senteces with the word god AND the word created
# NOTE that the concordance function does not return any thing but only prints the results to  
# the screen if we want to use these sentences for further analysis we can use 
# the concordance_list function 

my_concordance_list = text3.concordance_list(["god","created"]) 
print("\n\n=====\n\n")
print(f"my_concordance_list type : {type(my_concordance_list)}\n")
print(f"this is my concorance list:\n{my_concordance_list}")


```

```python
# we can use nltk.text.simmilar function to find words that frequently appear 
# in the context of a givven word 
# for exapmle we will look at the 20 most frequwent words that appeared in the context of 
# the word "god" in both our texts 
print(f"the 20 most frequent words in {text1.name} :")
print(text1.similar("god",20))
print("\n\n")
print(f"the 20 most frequent words in {text3.name} :")
print(text3.similar("god",20) )
```

# PreProcessing 


**Stemming and lemmatization:** Stemming is an informal process of converting words to their base forms using heuristic rules. For example, university, universities, and universitys might all be mapped to the base univers. (One limitation in this approach is that universe may also be mapped to univers, even though universe and university dont have a close semantic relationship.) Lemmatization is a more formal way to find roots by analyzing a words morphology using vocabulary from a dictionary. Stemming and lemmatization are provided by libraries like spaCy and NLTK. 

```python
from nltk.stem import WordNetLemmatizer
from nltk.stem.porter import PorterStemmer


lemmatizer = WordNetLemmatizer()
stemmer = PorterStemmer()
my_words = ["universe",
    "university", "universities", "universitying", "universityed",
    "universitying", "universal", "universe's", "universalize"
]
my_words_stemmed = []
my_words_lemmatized = []

for word in my_words:
    my_words_stemmed.append(stemmer.stem(word))
    my_words_lemmatized.append(lemmatizer.lemmatize(word))

print("Original      | Stemmed      | Lemmatized")
print("-" * 45)
for i in range(len(my_words)):
    print(f"{my_words[i]:<15} | {my_words_stemmed[i]:<15} | {my_words_lemmatized[i]}")
```

# sentiment analysis 

```python
import nltk
import random
from nltk.corpus import movie_reviews
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('movie_reviews')
nltk.download('stopwords')
nltk.download('wordnet')

# initialize lemmatizer
lemmatizer = WordNetLemmatizer()
# get list of stop words
stop_words = set(stopwords.words('english'))
documents = []
all_words = []
all_words_lemmed = []

for category in movie_reviews.categories():
    
    for fileid in movie_reviews.fileids(category):
    
        words = []
        for word in movie_reviews.words(fileid):
            
            word = word.lower()
            lemmed_word = lemmatizer.lemmatize(word)
            
            # do not add non-alphanumeric words and stopwords
            if word.isalnum() and word not in stop_words:
                words.append(word)
                all_words.append(word)
                all_words_lemmed.append(lemmed_word)
                
        documents.append((words, category))


random.shuffle(documents)
# create a frequency distribution of all preprocessed words in the corpus
all_words = nltk.FreqDist(all_words)
all_words_lemmed = nltk.FreqDist(all_words_lemmed)
word_features = list(all_words.keys())[:3000]
word_features_lemmed = list(all_words_lemmed.keys())[:3000]

# thi function checks which of the top 3000 words are present in a given document
def find_features(document, word_features):
    words = set(document)
    features = {}
    
    for word in word_features:
        features[word] = (word in words)
    
    return features

featuresets_unlemm = []
for (rev, category) in documents:
    features = find_features(rev, word_features)
    featuresets_unlemm.append((features, category))

featuresets_lemmed = []
for (rev, category) in documents:
    features = find_features(rev, word_features_lemmed)  
    featuresets_lemmed.append((features, category))  

```

```python

training_size = int(len(featuresets_unlemm) * 0.9)

# set that we'll train our classifier with
training_set_unlemm = featuresets_unlemm[:training_size]

# set that we'll test against.
testing_set_unlemm = featuresets_unlemm[training_size:]

classifier = nltk.NaiveBayesClassifier.train(training_set_unlemm)
# NOTE that naiveBayes accureacy is vary volitile and depends on the shuffle we did earlyer 
print("unleemed classifier accuracy percent:",(nltk.classify.accuracy(classifier, training_set_unlemm))*100)

classifier.show_most_informative_features(15)


training_set_lemmed = featuresets_lemmed[:training_size]

# set that we'll test against.
testing_set_lemmed = featuresets_lemmed[training_size:]

classifier = nltk.NaiveBayesClassifier.train(training_set_lemmed)
# NOTE that naiveBayes accureacy is vary volitile and depends on the shuffle we did earlyer 
print("lemmed classifier accuracy percent:",(nltk.classify.accuracy(classifier, testing_set_lemmed))*100)

classifier.show_most_informative_features(15)
```
